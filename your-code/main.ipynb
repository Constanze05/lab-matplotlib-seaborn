#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x1 = np.arange(0,10,1)     # Return evenly spaced values within a given interval numpy.arange([start, ]stop, [step, ]dtype=None)
x2 = np.linspace(0,5,10)   # Return evenly spaced numbers over a specified interval

plt.plot(x1, x2**2, label='quadratic')
plt.plot(x1, np.exp(x2), label='exponential') 

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("quadratic vs exponential plots")
plt.legend()
plt.show()
```
